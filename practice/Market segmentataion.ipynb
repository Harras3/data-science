```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('3.0example.csv')
```

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
x = data.copy()
```

```python
kmeans = KMeans(4)
kmeans.fit(x)
kmeans.inertia_
```

```python
clusters = x.copy()
clusters['clusters_pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'], c = clusters['clusters_pred'],cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
kmeans.inertia_
```

```python
wcss =[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss    
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('Numbers of cllusters')
plt.ylabel('Wcss')
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['clusters_pred'] = kmeans_new.fit_predict(x_scaled)
```

```python
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'], c = clusters_new['clusters_pred'],cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```
